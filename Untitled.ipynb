```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img=cv2.imread(
        'EyeScreening_4.3_2018-3-12_1755_OD.jpg',
        0)
```

```python
img
```

```python
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(200,200))
cl1 = clahe.apply(img)
plt.imshow(cl1, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
clahe2 = cv2.createCLAHE(clipLimit=5, tileGridSize=(10,10))
cl2 = clahe2.apply(img)
plt.imshow(cl2, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
clahe3 = cv2.createCLAHE(clipLimit=10, tileGridSize=(10,10))
cl3 = clahe3.apply(img)
plt.imshow(cl3, cmap = 'gray', interpolation = 'bicubic')
  # to hide tick values on X and Y axis
plt.show()
```

```python
a=1
b=-8*a
kernel = np.array([ [a,a,a],  
                    [a,  b,a],  
                    [a, a, a] ]) 
img_lap = cv2.filter2D(img,cv2.CV_8UC3 , kernel)
plt.imshow(img_lap, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
